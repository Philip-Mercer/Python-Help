<h2/> Grouping

```python

data = {'Company':['GOOG', 'GOOG', 'MSFT', 'MSFT', 'FB', 'FB'],
       'Person':['Sam', 'Charlie', 'Amy', 'Vanessa', 'Carl', 'Sarah'],
       'Sales':[200, 120, 340, 124, 243, 350]}
```

```python
df_sales = pd.DataFrame(data)
```

```python
df_sales
```

```python
byCompany = df_sales.groupby('Company')
```

```python
byCompany.mean(numeric_only=True)
```

```python
byCompany.sum(numeric_only=True)
```

```python
byCompany.std(numeric_only=True)
```

```python
byCompany.sum(numeric_only=True).loc['FB']
```

```python
# Alternatively ...

df_sales.groupby('Company').sum().loc['FB']
```

```python
df_sales.groupby('Company').count()
```

```python
df_sales.groupby('Company').max()
```

```python
df_sales.groupby('Company').min()
```

```python
df_sales.groupby('Company').describe()
```

```python
df_sales.groupby('Company').describe().transpose()
```

```python
df_sales.groupby('Company').describe().transpose()['FB']
```

```python
df_sales.groupby(['Company', 'Person']).describe() # Group by multiple fields with a list
```

<h2/> Merging

```python
df_mer1 = pd.DataFrame({'A': ['A0', 'A1', 'A2', 'A3'],
                       'B': ['B0', 'B1', 'B2', 'B3'],
                       'C': ['C0', 'C1', 'C2', 'C3'],
                       'D': ['D0', 'D1', 'D2', 'D3'],}, 
                      index=[0, 1, 2, 3])
```

```python
df_mer2 = pd.DataFrame({'A': ['A4', 'A5', 'A6', 'A7'],
                       'B': ['B4', 'B5', 'B6', 'B7'],
                       'C': ['C4', 'C5', 'C6', 'C7'],
                       'D': ['D4', 'D5', 'D6', 'D7'],}, 
                      index=[4, 5, 6, 7])
```

```python
df_mer3 = pd.DataFrame({'A': ['A8', 'A9', 'A10', 'A11'],
                       'B': ['B8', 'B9', 'B10', 'B11'],
                       'C': ['C8', 'C9', 'C10', 'C11'],
                       'D': ['D8', 'D9', 'D10', 'D11'],}, 
                      index=[8, 9, 10, 11])
```

```python
df_mer1
```

```python
df_mer2
```

```python
df_mer3
```

```python
pd.concat([df_mer1, df_mer2, df_mer3])
```

```python
pd.concat([df_mer1, df_mer2, df_mer3], axis=1)
```

```python
df_left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', ], 
            'A': ['A0', 'A1', 'A2', 'A3', ], 
            'B': ['B0', 'B1', 'B2', 'B3', ]})

df_right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3', ], 
            'C': ['C0', 'C1', 'C2', 'C3', ], 
            'D': ['D0', 'D1', 'D2', 'D3', ]})
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(df_left, df_right, how='inner', on='key')
```

```python
df_leftn = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2', ], 
            'key2': ['K0', 'K1', 'K0', 'K1', ],
            'A': ['A0', 'A1', 'A2', 'A3', ], 
            'B': ['B0', 'B1', 'B2', 'B3', ]})

df_rightn = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2', ], 
            'key2': ['K0', 'K0', 'K0', 'K0', ],
            'C': ['C0', 'C1', 'C2', 'C3', ], 
            'D': ['D0', 'D1', 'D2', 'D3', ]})
```

```python
df_leftn
```

```python
df_rightn
```

```python
pd.merge(df_leftn, df_rightn, on=['key1', 'key2'])
```

```python
pd.merge(df_leftn, df_rightn, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(df_leftn, df_rightn, on=['key1', 'key2'], how='right')
```

<h3/> Joining is done on the dataframe INDEX, not on the key fields defined in a MERGE

```python
df_join_left = pd.DataFrame({'A': ['A0', 'A1', 'A2'], 
                            'B': ['B0', 'B1', 'B2']}, 
                            index=['K0', 'K1', 'K2'])

df_join_right = pd.DataFrame({'C': ['C0', 'C1', 'C2'], 
                            'D': ['D0', 'D1', 'D2']}, 
                            index=['K0', 'K2', 'K3'])
```

```python
df_join_left.join(df_join_right)
```

```python
df_join_left.join(df_join_right, how='outer')
```
